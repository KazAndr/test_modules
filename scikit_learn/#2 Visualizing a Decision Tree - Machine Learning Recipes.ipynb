# Определение типа цветка на базе данных ирисов

```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn import tree
```

```python
iris = load_iris()
```

```python
print(iris.feature_names)
print(iris.target_names)
```

```python
print(iris.data[0])
```

Разделим данные на наборы для тестирования и обучения. Всего набор включает 150 измерений от 0 до 49 для сентозы, от 50 до 99 для разноцветного и от 100 до 149 для верджиники (по индексам.)

```python
test_idx = [0, 50, 100]
```

```python
train_target = np.delete(iris.target, test_idx)
train_data = np.delete(iris.data, test_idx, axis=0)
```

```python
test_target = iris.target[test_idx]
test_data = iris.data[test_idx]
```

```python
clf = tree.DecisionTreeClassifier()
```

```python
clf = clf.fit(train_data, train_target)
```

```python
clf.predict(test_data)
```

Метки указываемые классивикатором для тестовых данных точно соответствуют ожидаемым.

Визуализируем дерево решений

```python
import graphviz 
dot_data = tree.export_graphviz(clf, out_file=None, 
                     feature_names=iris.feature_names,  
                     class_names=iris.target_names,  
                     filled=True, rounded=True,  
                     special_characters=True)  
graph = graphviz.Source(dot_data)  
graph 
```
