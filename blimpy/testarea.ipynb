```python
from blimpy import Waterfall
import matplotlib.pyplot as plt
```

```python
voyadger_file = '../some_data_for_testing/filterbank_files/Voyager1.single_coarse.fine_res.h5'
frb_file = '../some_data_for_testing/filterbank_files/FRB180417.fil'
crab_data = '../some_data_for_testing/filterbank_files/2020_05_31_11_45_46.fil'
```

## Testing with Voyadger data

```python
observation = Waterfall(voyadger_file)
```

```python
%matplotlib inline
```

```python
observation.info()
```

```python
print(observation.header)
print(observation.data.shape)
```

```python
observation.plot_spectrum(logged=True)
```

```python
observation.plot_spectrum(f_start=8419.26, f_stop=8419.34)
```

```python
plt.figure(figsize=(8, 6))
plt.subplot(3,1,1)
observation.plot_spectrum(f_start=8419.274, f_stop=8419.275) # left sideband
plt.subplot(3,1,2)
observation.plot_spectrum(f_start=8419.296, f_stop=8419.298) # carrier
plt.subplot(3,1,3)
observation.plot_spectrum(f_start=8419.319, f_stop=8419.320) # right sideband
plt.tight_layout()
```

```python
observation.plot_waterfall()
```

```python
observation.plot_waterfall(f_start=8419.296, f_stop=8419.298)
```
